```python
!pip install pytubefix
import time
import os
from IPython.display import clear_output
from pytubefix import YouTube
print(" Bibliotecas instaladas com sucesso!")
```

```python
#defLogo
def logo():
  print("")
  print(" BLACK  HOLE ")
  print("")

#def tela principal
def tela ():
  print(f" URL: {link}")
  print(f" Vdeo: {media.title}")
  print(f" Canal: {media.author}")
  print(" ")
  print(" -- Download--")
  print(" 1 - Baixar Vdeo (MP4)")
  print(" 2 - Baixar udio (M4A)")

logo()
time.sleep(2)
stream = ""
select = ""

#controle de repetio
while select != "N" :
  clear_output()
  while select != 99:
    logo()
    link = input(" Digite a URL: ")
    try:
      media = YouTube(link)
      select = 99
    except:
      clear_output()
      print(" URL invlida")
      time.sleep(1)
      clear_output()

  #Tela principal
  clear_output()
  logo()
  tela()
  select = input(" Digite a opo: ")
  clear_output()

  #Tratativa de opo invalida
  while select != "1" and select != "2":
    clear_output()
    print(" Opo invlida")
    time.sleep(1)
    clear_output()
    logo()
    tela()
    select = input(" Digite a opo: ")

   #Seleo de formato
  if select == "1":
    stream = media.streams.get_highest_resolution()
    print(" ")
    print(" Download ...")
    time.sleep(1)
    stream.download()
    clear_output()
  elif select == "2":
    stream = media.streams.get_audio_only()
    print(" ")
    print(" Download ...")
    time.sleep(1)
    stream.download()
    clear_output()
  #Download Concludo
  logo()
  print(" Download concludo !")
  sel = input(" Deseja baixar outro Vdeo(S/N): ")
  select = sel.upper()
  clear_output()
```
