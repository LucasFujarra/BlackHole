```python
!pip install pytubefix
import time
import os
from IPython.display import clear_output
from pytubefix import YouTube
print("█ Bibliotecas instaladas com sucesso!")
```

```python
#defLogo
def logo():
  print("███████████████████████████████████")
  print("██████████ BLACK ֍ HOLE ███████████")
  print("███████████████████████████████████")

#def tela principal
def tela ():
  print(f"█ URL: {link}")
  print(f"█ Vídeo: {media.title}")
  print(f"█ Canal: {media.author}")
  print("█ ")
  print("█ -- Download--")
  print("█ 1 - Baixar Vídeo (MP4)")
  print("█ 2 - Baixar Áudio (M4A)")

logo()
time.sleep(2)
stream = ""
select = ""

#controle de repetição
while select != "N" :
  clear_output()
  while select != 99:
    logo()
    link = input("█ Digite a URL: ")
    try:
      media = YouTube(link)
      select = 99
    except:
      clear_output()
      print("█ URL inválida")
      time.sleep(1)
      clear_output()

  #Tela principal
  clear_output()
  logo()
  tela()
  select = input("█ Digite a opção: ")
  clear_output()

  #Tratativa de opção invalida
  while select != "1" and select != "2":
    clear_output()
    print("█ Opção inválida")
    time.sleep(1)
    clear_output()
    logo()
    tela()
    select = input("█ Digite a opção: ")

   #Seleção de formato
  if select == "1":
    stream = media.streams.get_highest_resolution()
    print("█ ")
    print("█ Download ...")
    time.sleep(1)
    stream.download()
    clear_output()
  elif select == "2":
    stream = media.streams.get_audio_only()
    print("█ ")
    print("█ Download ...")
    time.sleep(1)
    stream.download()
    clear_output()
  #Download Concluído
  logo()
  print("█ Download concluído !")
  sel = input("█ Deseja baixar outro Vídeo(S/N): ")
  select = sel.upper()
  clear_output()
```
